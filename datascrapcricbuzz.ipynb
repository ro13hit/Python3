```python
from bs4 import BeautifulSoup
import requests
url=requests.get("https://www.cricbuzz.com/cricket-match/live-scores")
soup=url.text
s=BeautifulSoup(soup,"html.parser")
for i in s.find_all('div',{'class':"cb-mtch-lst cb-col cb-col-100 cb-tms-itm"}):
    print(i.text)
    print('\n')
```
