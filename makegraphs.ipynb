```python
import matplotlib.pyplot as p
```

```python
l=[1,2,3,4,5,6]
n=[2,3,5,7,9,1]
```

```python
p.plot(l,n)
```

```python
p.xlabel('stocks')
```

```python
p.xlabel("Stocks")
p.ylabel("Units")
p.plot(l,n)
```

```python
p.title("STOCKS FOR RG")
p.xlabel("Stocks")
p.ylabel("Units")
p.plot(l,n)
```

```python
y=[9,8,7,6,5]
z=[2,5,9,3,7]
p.xlabel("Stocks")
p.ylabel("Units")
p.title('STOCKS COMPANY')
p.plot(l,color='Black',label='Year1',linewidth=5)
p.plot(n,color='Red',label='Year2',linewidth=4)
p.plot(y,color='Yellow',label='Year3',linewidth=3)
p.plot(z,label='Year4',linewidth=2)
p.legend()
```

```python
p.show()
```

```python
x=[1,2,3,4,5]
y=[6,5,4,3,2]
p.bar(x,y,color='Red',label='Age Group')
p.xlabel('Age')
p.ylabel('Units')
p.legend()
```

```python
x=[1,2,3,4,5]
y=[6,5,4,3,2]
a=[6,7,8,9,10]
b=[11,10,9,8,7]
p.xlim(0,13)
p.ylim(0,7)
p.bar(x,y,color='DarkOrange',label='Group1')
p.bar(6,2,color='LightBlue',label='Group0')
p.bar(b,y,color='Green',label='Group2')
p.xlabel('Age')
p.ylabel('Units')
p.legend()
```

```python

x=[1,2,3,4,5]
y=[6,5,4,3,2]
a=[6,7,8,9,10]
b=[11,10,9,8,7]
p.xlim(0,13)
p.ylim(0,7)
p.bar(x,y)
p.bar(6,2)
p.bar(b,y)
p.xlabel('Age')
p.ylabel('Units')
p.legend('BOB')
```
